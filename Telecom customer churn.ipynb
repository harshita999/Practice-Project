```python
x=df.drop('TotalCharges',axis=1)
y=df['TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
sc.fit(df)
sd=sc.transform(df)
sd
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
max_accuracy=0
max_randomstate=0
for i in range(1,500):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state=i)
    lr= LinearRegression()
    lr.fit(x_train, y_train)
    predlr = lr.predict(x_test)
    acc=r2_score(y_test, predlr)
    if acc>max_accuracy:
        max_accuracy=acc
        max_randomstate=i
print("Maximum r2 score is ",max_accuracy," on Random_state ",max_randomstate)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=478)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print("Accuracy score :  ", lr.score(x_train,y_train))
print("mean squared error : ", mean_squared_error(y_test,pred))
print("mean absolute error : ", mean_absolute_error(y_test,pred))
print("r2 Score : ", r2_score(y_test,pred))
```

# Regularization

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
# lets predict with Lasso first
ls=Lasso(alpha=0.01)
ls.fit(x_train,y_train)
pred=ls.predict(x_test)
print("Accuracy score : ",ls.score(x_train,y_train))
print("mean squared error : ", mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score: ",r2_score(y_test,pred))
```

```python
# Lets predict with Ridge now
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
pre=rd.predict(x_test)
print("Accuracy score : ",rd.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,pre))
print("mean absolute error : ",mean_absolute_error(y_test,pre))
print("r2 score : ",r2_score(y_test,pre))
```

```python
# now lastly lets predict Elasticnet
en=ElasticNet(alpha=0.0001)
en.fit(x_train,y_train)
preddd=en.predict(x_test)
print("Accuracy score : ",en.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,preddd))
print("mean absolute error : ",mean_absolute_error(y_test,preddd))
print("r2 score : ",r2_score(y_test,pred))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
print("Accuracy score : ",dtr.score(x_train,y_train))
print("mean square error : ",mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score :",r2_score(y_test,pred))
```

Decision Tree has 100% accuracy score and good R2 score 

# Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
pre=dtr.predict(x_test)
print("Accuracy score : ",gbr.score(x_train,y_train))
print("mean square error : ",mean_squared_error(y_test,pre))
print("mean absolute error : ",mean_absolute_error(y_test,pre))
print("r2 score :",r2_score(y_test,pre))
```

Gradient Boosting Regressor has accuracy score of 99.98%

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print("Accuracy score : ", rf.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score : ",r2_score(y_test,pred))
```

Random Forest egressor is giving good accuracy and fine R2 Score

# Cross Validation score

```python
from sklearn.model_selection import cross_val_score
```

```python
#linear regression
print('linear regression:',cross_val_score(lr,x,y,cv=5).mean())

#lasso
print('lasso:',cross_val_score(ls,x,y,cv=5).mean())

#Ridge
print('Ridge:',cross_val_score(rd,x,y,cv=5).mean())

#ElasticNet
print('ElasticNet:',cross_val_score(en,x,y,cv=5).mean())

#Decision Tree Regressor
print('Decision Tree Regressor:',cross_val_score(dtr,x,y,cv=5).mean())

#Gradient Boosting Regressor
print('Gradient Boosting Regressor:',cross_val_score(gbr,x,y,cv=5).mean())

#Random Forest Regressor
print('Random Forest Regressor:',cross_val_score(rf,x,y,cv=5).mean())
```

Random Forest Regressor to be the best model with accuracy score of 99.9%
