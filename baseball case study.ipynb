```python
x=features
y=df1["W"]
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif(x):
    vif=pd.DataFrame()
    vif['VARIABLES']=x.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
```

```python
vif(x)
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(sdf1,y3,test_size=.25,random_state=40)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print("Accuracy score :  ", lr.score(x_train,y_train))
print("mean squared error : ", mean_squared_error(y_test,pred))
print("mean absolute error : ", mean_absolute_error(y_test,pred))
print("r2 Score : ", r2_score(y_test,pred))
```

```python
#linear regression with min-max scaled
x_train,x_test,y_train,y_test=train_test_split(d,y1,test_size=.33, random_state=42)
```

```python
lrr=LinearRegression()
lrr.fit(x_train,y_train)
predd=lrr.predict(x_test)
print("Accuracy score : ", lrr.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,predd))
print("mean absolute error : ",mean_absolute_error(y_test,predd))
print("r2 score : ",r2_score(y_test,predd))
```

# Regularization

```python
# l1 and l2 Regularization and ElasticNet
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
x_train,x_test,y_train,y_test=train_test_split(d,y1,test_size=.20,random_state=3)
```

```python
# lets predict with Lasso first
ls=Lasso(alpha=0.01)
ls.fit(x_train,y_train)
pred=ls.predict(x_test)
print("Accuracy score : ",ls.score(x_train,y_train))
print("mean squared error : ", mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score: ",r2_score(y_test,pred))
```

```python
# Lets predict with Ridge now
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
pre=rd.predict(x_test)
print("Accuracy score : ",rd.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,pre))
print("mean absolute error : ",mean_absolute_error(y_test,pre))
print("r2 score : ",r2_score(y_test,pre))
```

```python
# now lastly lets predict Elasticnet
en=ElasticNet(alpha=0.0001)
en.fit(x_train,y_train)
preddd=en.predict(x_test)
print("Accuracy score : ",en.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,preddd))
print("mean absolute error : ",mean_absolute_error(y_test,preddd))
print("r2 score : ",r2_score(y_test,pred))
```

Ridge Regression is giving the best scores out of Lasso, Ridge and ElasticNet

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
print("Accuracy score : ",dtr.score(x_train,y_train))
print("mean square error : ",mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score :",r2_score(y_test,pred))
```

Decision Tree has 100% accuracy score and good R2 score of 89.9%

# Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
pre=dtr.predict(x_test)
print("Accuracy score : ",gbr.score(x_train,y_train))
print("mean square error : ",mean_squared_error(y_test,pre))
print("mean absolute error : ",mean_absolute_error(y_test,pre))
print("r2 score :",r2_score(y_test,pre))
```

Gradient Boosting Regressor has accuracy score of 99.99%

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print("Accuracy score : ", rf.score(x_train,y_train))
print("mean squared error : ",mean_squared_error(y_test,pred))
print("mean absolute error : ",mean_absolute_error(y_test,pred))
print("r2 score : ",r2_score(y_test,pred))
```

Random Forest egressor is giving good accuracy and fine R2 Score
